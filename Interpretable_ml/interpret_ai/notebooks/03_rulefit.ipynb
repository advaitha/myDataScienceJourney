```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn import metrics
from imodels import FIGSClassifier
from sklearn.datasets import load_breast_cancer
import os
```

```python
raw_data = load_breast_cancer()
```

```python
X = raw_data.data
y = raw_data.target
features = raw_data.feature_names
target_names = raw_data.target_names
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.20,random_state=123)
```

```python
figs = FIGSClassifier(max_rules=4)
```

```python
figs.fit(X_train,y_train,feature_names=features)
```

```python
print(figs)
```

```python
print(figs.print_tree(X_train, y_train))
```

```python
figs.plot(fig_size=8)
```
